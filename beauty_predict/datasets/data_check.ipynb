```python
from PIL import Image
from tqdm import tqdm
```

Какие размеры картинок бывают:

```python
img_size_df = {"rating": [], "id": [], "height": [], "width": []}
root_dir = "all_data"
for subdir in ["0", "1", "2", "3"]:
    sub_path = os.path.join(root_dir, subdir)
    for img_name in tqdm(os.listdir(sub_path)):
        img = Image.open(os.path.join(sub_path, img_name))
        img_size_df["rating"].append(subdir)
        img_size_df["id"].append(img_name.split(".")[0])
        img_size_df["width"].append(img.size[0])
        img_size_df["height"].append(img.size[1])
```

```python
data_checkimg_size_df = pd.DataFrame(img_size_df)
```

```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect('core.download.db')

# Получение списка таблиц в базе данных
query = "SELECT name FROM sqlite_master WHERE type='table';"
tables = pd.read_sql(query, conn)

# Вывод списка таблиц
print(tables)

# Закрытие соединения
conn.close()
```

```python

# Установка соединения с базой данных SQLite
conn = sqlite3.connect('core.download.db')

# Выполнение SQL-запроса и чтение результата в DataFrame
query = "SELECT * FROM img_info"
df = pd.read_sql(query, conn)

# Закрытие соединения
conn.close()

# Вывод содержимого DataFrame
print(df.shape)
```

```python

# Установка соединения с базой данных SQLite
conn = sqlite3.connect('core.download.db')

# Выполнение SQL-запроса и чтение результата в DataFrame
query = "SELECT * FROM metadata"
meta_df = pd.read_sql(query, conn)

# Закрытие соединения
conn.close()

# Вывод содержимого DataFrame
print(meta_df.head())
```

```python
rated_df = df[df.id.isin(meta_df.id)]
rated_df.shape
```

```python
rated_df
```

```python
merged_df = rated_df.merge(meta_df, on='id', how='left')
```

```python
merged_df.head()
```

```python
merged_df[["id", "img_path", "img_name", "rating"]].to_csv("nessesary_files.csv")
```

```python
import pandas as pd
import json
import re
import string 
```

```python
ann_df = pd.read_csv("ann.csv").drop("Unnamed: 0", axis=1)
ann_df.columns

ann_df["rating"] = ann_df["rating"] - 1
```

```python
ann_df["rating"].unique()
```

```python
def extract_text(string):
    pattern = r'\d+_(.*)' 
    match = re.match(pattern, string)
    
    if match:
        return match.group(1)  
    else:
        return None 

def rm_dupls_punctuation(line):
    result = ""
    prev_char = None

    for char in line:
        if char not in string.punctuation:
            result += char
            prev_char = char
        else:
            if char != prev_char:
                result += char
                prev_char = char

    return result

signs = {i: 0 for i in string.punctuation}
```

```python
classes = {}
for i in range(4):
    class_df = ann_df[ann_df["rating"] == i][["id", "img_generation_promt"]]
    classes[i] = class_df.values.tolist()
    for k, _ in enumerate(classes[i]):
        promt = _[1]
        for c in promt:
            if c in signs.keys():
                signs[c] += 1
        classes[i][k][1] = rm_dupls_punctuation(
                                    extract_text(promt)
                                ).replace("_", " ").strip() 
classes
```

```python
full_set = {
    "root_dir": "all_data",
    "classes": classes
}
```

```python
with open("full_set.json", "w") as f:
    json.dump(full_set, f, indent=4)
```
